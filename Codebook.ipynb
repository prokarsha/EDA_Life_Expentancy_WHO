```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Life Expectancy Data.csv')
```

```python
# Visualization before filling missing values
sns.heatmap(data.isnull(), cmap='viridis', cbar=False)
plt.title('Heatmap')
plt.show()
```

```python
# Fill missing values with zero
data.fillna(0, inplace=True)
```

```python
# Visualization after filling missing values
sns.heatmap(data.isnull(), cmap='viridis', cbar=False)
plt.title('Heatmap')
plt.show()
```

```python
# Histogram of Life Expectancy
plt.figure(figsize=(10, 6))
sns.histplot(data['Life expectancy '], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Life Expectancy')
plt.xlabel('Life Expectancy')
plt.ylabel('Frequency')
plt.show()
```

```python
# Boxplot of Life Expectancy by Status
plt.figure(figsize=(10, 6))
sns.boxplot(x='Status', y='Life expectancy ', data=data, palette='pastel')
plt.title('Life Expectancy by Status')
plt.xlabel('Status')
plt.ylabel('Life Expectancy')
plt.show()
```

```python
# Scatterplot of Life Expectancy vs GDP
plt.figure(figsize=(10, 6))
sns.scatterplot(x='GDP', y='Life expectancy ', data=data, hue='Status', palette='viridis')
plt.title('Life Expectancy vs GDP')
plt.xlabel('GDP')
plt.ylabel('Life Expectancy')
plt.show()
```

```python
# Pairplot of selected attributes
attributes = ['Life expectancy ', 'Adult Mortality', 'Alcohol', 'GDP']
sns.pairplot(data[attributes])
plt.show()
```
